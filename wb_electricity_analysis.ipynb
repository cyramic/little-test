```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('wb_electricity_data.csv')
data.head()
```

```python
# Let's look at Angola first
data_angola = data[data['Country Code'] == 'AGO']
data_angola.head()
```

```python
# Let's drop some of the columns we don't need
data_angola = data_angola.drop(['Country Name', 'Country Code', 'Series Code', '2018 [YR2018]'], axis=1)
data_angola.head()
```

```python
#Let's melt and pivot (below) the table to make it nicer
data_angola = pd.melt(data_angola, id_vars = 'Series Name', var_name = 'Year', value_name = 'Value')
data_angola['Year'] = data_angola.Year.str[0:4]
data_angola.head()
```

```python
data_angola = data_angola.pivot(index='Year', columns = 'Series Name', values = 'Value')
```

```python
data_angola.plot()
plt.ylabel('% of total population')
plt.xticks(rotation=45)
plt.show()
```
